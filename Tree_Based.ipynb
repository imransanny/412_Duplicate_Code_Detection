```python
import ast
import networkx as nx

def parse_file_to_ast(file_path):
    with open(file_path, "r") as file:
        code = file.read()
    return ast.parse(code)

def ast_to_graph(node, graph=None, parent=None, edge_label=''):
    if graph is None:
        graph = nx.DiGraph()

    graph.add_node(id(node), label=type(node).__name__)

    if parent is not None:
        graph.add_edge(id(parent), id(node), label=edge_label)

    for field, value in ast.iter_fields(node):
        if isinstance(value, list):
            for item in value:
                if isinstance(item, ast.AST):
                    ast_to_graph(item, graph, node, field)
        elif isinstance(value, ast.AST):
            ast_to_graph(value, graph, node, field)

    return graph

def compare_graphs(graph1, graph2):
    return nx.is_isomorphic(graph1, graph2, node_match=lambda n1, n2: n1['label'] == n2['label'])

file_path = "/content/write a simple code for testing.py"
ast1 = parse_file_to_ast(file_path)
ast2 = parse_file_to_ast(file_path)
graph1 = ast_to_graph(ast1)
graph2 = ast_to_graph(ast2)

if compare_graphs(graph1, graph2):
    print("Duplicate code detected.")
else:
    print("No duplicates found.")
```
